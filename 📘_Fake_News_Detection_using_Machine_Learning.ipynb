```python
import pandas as pd
import numpy as np
```

```python
true = pd.read_csv('True.csv')
```

```python
fake = pd.read_csv('Fake.csv')
```

```python
true.head()
```

```python
fake.head()
```

```python
true['label'] = 1
```

```python
fake['label'] = 0
```

```python
news = pd.concat([fake, true], axis = 0)
```

```python
news.head()
```

```python
news.tail()
```

```python
news = news.drop(['title', 'subject', 'date'], axis = 1)
```

```python
news.head()
```

```python
news = news.sample(frac=1)                          #Reshuffle
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'], axis=1, inplace = True)
```

```python
news.head()
```

```python
import re
import string
```

```python
def wordopt(text):
# Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove digits
    text = re.sub(r'\d+', '', text)

    # Remove newlines and excess whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(xv_train, y_train)
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train, y_train)
```

```python
pred_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(xv_train, y_train)
```

```python
pred_gbc = gbc.predict(xv_test)
```

```python
gbc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gbc))
```

```python
def output_label(n):
  if n == 0:
    return "It is a fake news"
  elif n == 1:
    "It is a Genuine News"
```

```python
def manual_testing(news):
    # Correct dictionary syntax
    testing_news = {"text": [news]}

    # Create DataFrame
    new_def_test = pd.DataFrame(testing_news)

    # Apply text preprocessing (assumes wordopt() is defined)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)

    # Extract cleaned text
    new_x_test = new_def_test["text"]

    # Vectorize using fitted vectorizer (assumes vectorization is fitted already)
    new_xv_test = vectorization.transform(new_x_test)

    # Make predictions
    pred_lr = LR.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)

    # Return human-readable results (assumes output_label() is defined)
    return "\n\nLR Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(
        output_label(pred_lr[0]), output_label(pred_gbc[0]), output_label(pred_rfc[0]))
```

```python
news_article = str(input())
```

```python
manual_testing(news_article)
```
